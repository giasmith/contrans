### Inner Join 
Inner join deletes everything that does not have things in both sets

```python
myquery = '''
SELECT *
FROM nfl
INNER JOIN nba
    ON nfl.city = nba.city

'''
pd.read_sql(myquery, sportsteams)
```

### Left Join
Will prompt everything after 'FROM' and will make the relationships with other table even if there isn't a match  
Example: some NBA teams do not have NFL teams in their city so there are 'None' placed in their spaces

```python
myquery = '''
SELECT *
FROM nfl f
LEFT JOIN nba b
    ON f.city = b.city

'''
pd.read_sql(myquery, sportsteams)
```

### Cross Join
List all possible combinations of the data in both tables

```python
myquery = '''
SELECT *
FROM team_games t
INNER JOIN player_game pg
    ON t.game_id = pg.game_id
INNER JOIN players p
    ON pg.player_id = p.player_id

'''
pd.read_sql(myquery, nba_db)
```

### Ordering the players by their points (greatest-to-least)

```python
myquery = '''
SELECT 
    p.player AS player_name,
    t.game_id AS game,
    t.Team_Abbrev AS team, 
    pg.pts AS points
FROM team_games t
INNER JOIN player_game pg
    ON t.game_id = pg.game_id
INNER JOIN players p
    ON pg.player_id = p.player_id
WHERE p.player LIKE 'j% k%'
ORDER BY points DESC
LIMIT 10 OFFSET 5
'''
pd.read_sql(myquery, nba_db)
```
