## Second Normal Form
Every non-prime column must depend on the ENTIRE primary key (gameid + playerid) and not just part of the primary key (not just the playerid or gameid)

```python
nba['game_player_id']=nba['game_id']+nba['player_id']
```

## Third Normal Form
Need all of the dependecies of a second normal form

Calculated column:
-fg_pct
-fg
-fg3_pct
We'll delete these

Transitive Dependencies: 
-some columns depend depend on player
-some columns depend on game
-some columns depend on team
-some columns depend on team+game
Each of these will get a seperate table!
#### Among the non-primes we can't have any dependencies

```python
nba = nba.drop('fg_pct','fg','fg3_pc')
```

```python
nba_team_game = nba[['Team_Abbrev', 'Opponent_Abbrev', 'game_id', 'H_A', 
                     'Team_Score', 'Team_pace', 'Team_efg_pct', 'Team_tov_pct', 
                     'Team_orb_pct','Team_ft_rate', 'Team_off_rtg',]]
nba_team_game=nba_team_game.drop_duplicates()
nba_team_game.head(20)
```

```python
nba_game = nba[['game_id', 'game_date','OT']]
nba_game=nba_game.drop_duplicates()
nba_game.head(10)
```

```python
nba_player = nba[['player', 'player_id']]
nba_player = nba_player.drop_duplicates()
nba_player
```

```python
nba_player_game = nba.drop(['player', 'player_id','game_date','OT', 'Team_Abbrev', 
                            'Opponent_Abbrev', 'H_A', 'Team_Score', 'Team_pace', 
                            'Team_efg_pct', 'Team_tov_pct', 'Team_orb_pct','Team_ft_rate',
                            'Team_off_rtg'], axis=1)
nba_player_game
```

```python
nba_db =sqlite3.connect("nba.db")
```

```python
# NBA games table in nba_db
nba_game.to_sql('games', nba_db, index=False, chunksize=1000, if_exists='replace')
nba_player_game.to_sql('player_game', nba_db, index=False, chunksize=1000, if_exists='replace')
nba_player.to_sql('players', nba_db, index=False, chunksize=1000, if_exists='replace')
nba_team_game.to_sql('team_games', nba_db, index=False, chunksize=1000, if_exists='replace')
```

```python
myquery = '''
SELECT *
FROM team_games
WHERE Team_Abbrev = 'CLE'
'''

pd.read_sql(myquery, nba_db)
```
